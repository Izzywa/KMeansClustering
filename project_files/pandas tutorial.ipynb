```python
import matplotlib.pyplot as plt
import pandas as pd


#this reads a csv file, storing into a var
data = pd.read_csv('happyScore_income.csv')
```

```python
# this prints out all the column names
print(data.columns)
```

```python
# this stores the columns in a variable
happy = data.happyScore
avg_income = data.avg_income
```

```python
# plot the graph with the income as the independent var,
# happiness as the depenedent var

plt.xlabel('average income')
plt.ylabel('happiness index')
plt.scatter(avg_income, happy, cmap='coolwarm', c=happy)
plt.show()
```

```python
import numpy as np
income_inequality = data.income_inequality
GDP = data.GDP
median_income = data.median_income
```

```python
# create new array
new_array = np.vstack((GDP, income_inequality, median_income)).T
col_min = np.min(new_array, 0)
col_max = np.max(new_array, 0)
# normalise the array
normed_array = (new_array - col_min) / (col_max - col_min)
```

```python
# scatter plot, not normalised
plt.scatter(GDP, income_inequality, s=(median_income / 15), cmap='coolwarm', c=median_income)
plt.show()
```

```python
#scatter plot, normalised
area = normed_array[:,2]* 1000
plt.xlabel('GDP')
plt.ylabel('Income Inequality')
plt.scatter(normed_array[:,0], normed_array[:,1], s=area, alpha=0.75, 
            cmap='coolwarm', c=normed_array[:,2])
plt.colorbar(label="Median Income")
plt.show()
```

```python
# creating K-Means clusters
from sklearn.cluster import KMeans
income = data['avg_income']
happy = data['happyScore']

# create a 2D array
income_happy = np.column_stack((income,happy))
# create a KMean analysis
km_res = KMeans(n_clusters=3).fit(income_happy)
# find the coordinates of the clusters
clusters = km_res.cluster_centers_

# plot the graph
plt.scatter(income, happy)
plt.scatter(clusters[:,0], clusters[:,1], alpha=0.5,
            s=1000)
plt.show()
```

```python
cluster_group = km_res.predict(income_happy)
# maps the index of cluster groups to each data
```

```python
# create new table with each row pointing to the cluster group
new_income_happy = np.column_stack((income,happy, cluster_group))
```

```python
# color the points with the corresponding clusters
# dont really need to make a new array, just pass the cluster group
# result into the c argument

fig, ax = plt.subplots()
scatter = ax.scatter(income_happy[:,0], income_happy[:,1],
            c=cluster_group, cmap="cool")

k_scatter = ax.scatter(clusters[:,0], clusters[:,1], alpha=0.5,
            s=1000, c=np.arange(len(clusters[:,0])), cmap="cool")

#create legends
plt.show()
```
