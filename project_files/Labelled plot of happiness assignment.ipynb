```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('happyScore_income.csv')
```

```python
data.columns
```

```python
# sort by the median income
copy_data = data.copy()
copy_data.sort_values("median_income", inplace=True)
copy_data
```

```python
# normalise
happy_max = copy_data['happyScore'].max()
happy_min = copy_data['happyScore'].min()
copy_data['happyScore'] = (copy_data['happyScore'] - happy_min)/(happy_max-happy_min)
```

```python
# get the happiest countries
happiest = copy_data[ copy_data['happyScore'] > 0.5 ]
happiest
```

```python
# normalise the avg satisfaction
satisfaction_max = happiest['avg_satisfaction'].max()
satisfaction_min = happiest['avg_satisfaction'].min()

happiest['avg_satisfaction'] = (happiest['avg_satisfaction'] - satisfaction_min)/(satisfaction_max - satisfaction_min)

happiest
```

```python
# plot 
plt.xlabel('Median Income')
plt.ylabel('Happiness')
plt.scatter(happiest['median_income'], happiest['happyScore'],
            s=happiest['avg_satisfaction']*1000,
            alpha=0.7,
            c=happiest['avg_satisfaction'],
            cmap="coolwarm")

plt.text(happiest.iloc[0]['median_income'],
         happiest.iloc[0]['happyScore'],
         round(happiest.iloc[0]['avg_satisfaction'],2))

plt.text(happiest.iloc[-1]['median_income'],
         happiest.iloc[-1]['happyScore'],
         round(happiest.iloc[-1]['avg_satisfaction'],2))
plt.colorbar(label='Average satisfaction')
plt.show()
```
